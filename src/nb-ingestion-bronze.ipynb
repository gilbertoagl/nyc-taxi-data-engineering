```python
import requests
import os
from notebookutils import mssparkutils

# Variables
TAXI_TYPE = "yellow" 
DATE_ID = "2023-01"
```

```python
# LÓGICA DE DESCARGA

# Construcción de rutas
FILE_NAME = f"{TAXI_TYPE}_tripdata_{DATE_ID}.parquet"
URL_FUENTE = f"https://d37ci6vzurychx.cloudfront.net/trip-data/{FILE_NAME}"

SAVE_PATH = f"/lakehouse/default/Files/raw_Data/{TAXI_TYPE}_Taxi/"
FULL_PATH = os.path.join(SAVE_PATH, FILE_NAME)

# Crear carpeta
if not os.path.exists(SAVE_PATH):
    os.makedirs(SAVE_PATH)

print(f"⬇Descargando: {URL_FUENTE}")
print(f"Guardando en: {FULL_PATH}")

# Descarga
response = requests.get(URL_FUENTE)

if response.status_code == 200:
    with open(FULL_PATH, "wb") as f:
        f.write(response.content)
    print("Exito.")
    mssparkutils.notebook.exit("OK")
else:
    raise Exception(f"Error {response.status_code}: No se pudo descargar {FILE_NAME}")
```
